```julia
import Polynomials
import Plots
n=20
f(x)=exp(sin(2**x))*(cos(2**x))^2-exp(sin(2**x))*(sin(2**x))
equidistant_nodes = LinRange(0,1,n+1)
p=Polynomials.fit(equidistant_nodes, f.(equidistant_nodes))
x=LinRange(0,1,1000)
Plots.plot(x,[f.(x),p.(x)])
```

```julia
import Polynomials
import Plots
x=Polynomials.Polynomial([0,1])

firstintegral=Polynomials.integrate(p(x))
secondintegral=Polynomials.integrate(firstintegral((x)))
# println(secondintegral(1))
u = secondintegral-secondintegral(1)*x
x=LinRange(0,1,1000)
Plots.plot(x,u.(x))
```
